```python
import pandas as pd

# 1. Load dataset belum berlabel
df_unlabeled = pd.read_csv("./CPMK_2/dataset_youtube_comment.csv")

# 2. Ambil hanya kolom Comment
df_unlabeled = df_unlabeled[["Comment"]].copy()

# 3. Acak datanya
df_unlabeled_shuffled = df_unlabeled.sample(frac=1, random_state=42).reset_index(drop=True)

# 4. Ambil 250 data pertama
df_unlabeled_250 = df_unlabeled_shuffled.head(250).copy()

# 5. Simpan sebagai CSV baru
df_unlabeled_250.to_csv("unlabeled_comment_250_shuffled.csv", index=False)

print("Selesai! File disimpan sebagai unlabeled_comment_250_shuffled.csv")
```

```python
import pandas as pd

# 1. BACA DATASET BERLABEL (dengan delimiter ;)
df_labeled = pd.read_csv("./Labeling_Data/dataset_labeled_250.csv", delimiter=";")

# Ubah agar dataset tidak menggunakan delimiter (pakai format CSV standar)
# Ini membuat df_labeled menjadi DataFrame biasa tanpa karakter pemisah dalam cell
df_labeled = df_labeled[["Comment", "Label"]].copy()

# 2. BACA DATASET BELUM BERLABEL
df_unlabeled = pd.read_csv("./unlabeled_comment_250_shuffled.csv")

# Ambil hanya kolom Comment
df_unlabeled = df_unlabeled[["Comment"]].copy()

# Ambil hanya 250 data pertama
df_unlabeled_250 = df_unlabeled.head(250).copy()

# 3. Pindahkan label dari dataset berlabel ke dataset belum berlabel
df_merged = df_unlabeled_250.merge(
    df_labeled,
    on="Comment",
    how="left"   # agar tetap muncul meskipun tidak ada label
)

# 4. SIMPAN HASILNYA KEMBALI DENGAN DELIMITER ;
output_path = "./dataset_unlabeled_with_label.csv"
df_merged.to_csv(output_path, sep=";", index=False)

print("Dataset berhasil disimpan ke:", output_path)
```
